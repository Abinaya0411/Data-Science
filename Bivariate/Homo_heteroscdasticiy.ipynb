```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
dependent=dataset[['Profit']]
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Standardisation for increasing the r2 score, for making the model as much as best
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)#for findout mu value and sigma value, standardised inputs assigned to xtrain
X_test=sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=100,random_state=10,criterion='mae')
regressor=regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
import numpy as np
residuals = np.array(y_test) - y_pred
```

```python
residuals
```

```python
import matplotlib.pyplot as plt
try:
    # Plot residuals against predicted values
    plt.figure(figsize=(4,2))
    plt.scatter(y_pred, residuals, color='blue')
    plt.axhline(y=0, color='red', linestyle='--')
    plt.xlabel('Predicted Values')
    plt.ylabel('Residuals')
    plt.title('Residuals vs Predicted Values')
    plt.show()
except Exception as e:
    print("Error occurred while plotting residuals:", e)
```

# Heteroscadasticity
The above code might be Heteroscdasticity because they are not in the same pattern

```python
#r2score evaluation metrics
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```
